```python

import pandas as pd
import numpy as np

df = pd.read_csv("./diabetes.csv")

def generate_realistic_data(df, n_samples=10000):
    synthetic_rows = []

    for outcome in [0, 1]:
        subset = df[df["Outcome"] == outcome]

        mean = subset.drop("Outcome", axis=1).mean().values
        cov = np.cov(subset.drop("Outcome", axis=1).values, rowvar=False)

        samples = np.random.multivariate_normal(mean, cov, size=n_samples // 2)

        for row in samples:
            row_dict = {}

            for i, col in enumerate(subset.columns[:-1]):
                value = row[i]

                # enforce realistic constraints
                if col in ["Glucose", "BloodPressure", "BMI"]:
                    value = max(0, value)

                row_dict[col] = float(value)

            row_dict["Outcome"] = outcome
            synthetic_rows.append(row_dict)

    synthetic_df = pd.DataFrame(synthetic_rows)

    # Merge with original
    final_df = pd.concat([df, synthetic_df], ignore_index=True)
    final_df.to_csv("./diabetes_generated.csv", index=False)

    print("Final dataset shape:", final_df.shape)
    return final_df


df_expanded = generate_realistic_data(df)
```
